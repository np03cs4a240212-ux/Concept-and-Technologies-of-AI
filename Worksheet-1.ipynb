```python
import numpy as np

matrix_C = np.array([[3,4],[7,8]])
matrix_D = np.array([[5,3],[2,1]])

# 1. Prove A * A^-1 = I
inverse_matrix_C = np.linalg.inv(matrix_C)
identity_check = np.dot(matrix_C, inverse_matrix_C)
print("Matrix C * Matrix C^-1 =\n", identity_check)

# 2. Prove AB ≠ BA
product_CD = np.dot(matrix_C, matrix_D)
product_DC = np.dot(matrix_D, matrix_C)
print("Matrix C * Matrix D =\n", product_CD)
print("Matrix D * Matrix C =\n", product_DC)

# 3. Prove (AB)^T = B^T A^T
transpose_product = np.transpose(product_CD)
product_transposes = np.dot(np.transpose(matrix_D), np.transpose(matrix_C))
print("(Matrix C * Matrix D)^T =\n", transpose_product)
print("Matrix D^T * Matrix C^T =\n", product_transposes)
```

```python
#• Solve the following system of Linear equation using Inverse Methods.

coefficient_matrix = np.array([[2, -3, 1],
                              [1, -1, 2],
                              [3,  1, -1]])

constant_vector = np.array([-1, -3, 9])
inverse_coefficient_matrix = np.linalg.inv(coefficient_matrix)
solution_vector = np.dot(inverse_coefficient_matrix, constant_vector)
print("Solution [x, y, z] =", solution_vector)
```

```python
import numpy as np
import time
```

```python
#1. Element-wise Addition

# Create two lists of size 1,000,000
first_number_list = list(range(1000000))
second_number_list = list(range(1000000))

start_time = time.time()
sum_result_list = [num1 + num2 for num1, num2 in zip(first_number_list, second_number_list)]
end_time = time.time()

print("Python list addition time:", end_time - start_time, "seconds")
```

```python
first_number_array = np.arange(1000000)
second_number_array = np.arange(1000000)

start_time = time.time()
sum_result_array = first_number_array + second_number_array
end_time = time.time()

print("NumPy array addition time:", end_time - start_time, "seconds")
```

```python
start_time = time.time()
product_result_array = first_number_array * second_number_array
end_time = time.time()

print("NumPy array multiplication time:", end_time - start_time, "seconds")
```

```python
#3. Dot Product

start_time = time.time()
dot_product_list = sum(num1 * num2 for num1, num2 in zip(first_number_list, second_number_list))
end_time = time.time()

print("Python list dot product time:", end_time - start_time, "seconds")
```

```python
start_time = time.time()
dot_product_array = np.dot(first_number_array, second_number_array)
end_time = time.time()

print("NumPy dot product time:", end_time - start_time, "seconds")
```

```python
#4. Matrix Multiplication (1000 × 1000)

# Create two 1000x1000 matrices using lists
first_matrix_list = [[i for i in range(1000)] for _ in range(1000)]
second_matrix_list = [[i for i in range(1000)] for _ in range(1000)]

start_time = time.time()
product_matrix_list = [[sum(element1 * element2 for element1, element2 in zip(row, column))
                  for column in zip(*second_matrix_list)]
                  for row in first_matrix_list]
end_time = time.time()

print("Python list matrix multiplication time:", end_time - start_time, "seconds")
```
